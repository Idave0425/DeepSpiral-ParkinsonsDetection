```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Set the directory where the files are located (Healthy)
directory_src = '/content/drive/Shareddrives/1:1_Ishan_Dave/Dataset/hw_dataset/hw_dataset/control'
directory_dest = '/content/drive/Shareddrives/1:1_Ishan_Dave/Dataset/Master_Data/Healthy'
# Create an empty dictionary to store dataframes
dataframes = {}

# Iterate over the files in the directory
for filename in os.listdir(directory_src):
    if filename.endswith('.txt'):  # Check if the file is a .txt file
      df = pd.read_csv(directory_src + '/' + filename, delimiter=';', header=None)
      df.columns = ['X', 'Y', 'Z', 'Pressure', 'GripAngle', 'Timestamp', 'Test ID']

      # Filter based on the Test ID if needed, for example Test ID 0
      df_test0 = df[df['Test ID'] == 0]
      # Plotting
      plt.figure(figsize=(10, 10))  # You can adjust the size of the image
      plt.plot(df_test0['X'], df_test0['Y'], marker='o')  # You can change to plt.scatter if that's preferred
      plt.axis('equal')  # To maintain aspect ratio
      # Save the plot as an image
      # Replace 'your_save_directory' with the directory you want to save the image in
      plt.savefig(directory_dest + '/' + filename[:-3] + 'png')  # Or use .jpg if you prefer
```
