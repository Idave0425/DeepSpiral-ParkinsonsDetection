### Example with random data

```python
#example on multiline plot
epochs = [20,30]
lr = [0.0001,0.001,0.005,0.01,0.1]
A = [0.9868,0.93654,0.879568,0.968745,0.89655]#20 epochs
B = [0.7852,0.89652,0.79652,0.7852,0.8125]#30 epochs


plt.figure(figsize = (8,8))
plt.plot(lr, A, "*-", label = f"Epochs: {epochs[0]}")
plt.plot(lr, B, "*-", label = f"Epochs: {epochs[1]}")
plt.xlabel("Learning Rate")
plt.ylabel("Validation Accuracy")
plt.xscale("log")
plt.title("Validation Accuracy with respective to learning rate and epochs")
plt.legend(loc = "best")
plt.grid(True)
plt.show()
```

### Using the sheet which is used to record the hyper parameter values

- Download the google sheet as a csv file.
- Upload it to the colab
- Copy the path and generate the graphs

```python
import pandas as pd
path = "sample_hyper_parameter_tuning_sheet.csv"
data = pd.read_csv(path)
data.head()
```

```python
epochs = list(data["Epochs/learning rate"])
lr = list(data.columns)
lr = lr[1:]
all_acc = []
for index in range(len(epochs)):
  all_acc.append(list(data.iloc[index,1:]))

plt.figure(figsize = (9,9))
for index in range(len(epochs)):
  plt.plot(lr, all_acc[index], label = "Epochs: {}".format(epochs[index]))
plt.xlabel("Learning Rate")
plt.ylabel("Accuracy")
plt.title("Accuracy values with Learning rate and Epochs")
plt.grid(True)
plt.legend(loc = "best")
plt.show()
```
